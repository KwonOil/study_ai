```python
# 상관계수 절대값이 0.5이상인 것들의 열 이름을 리스트로 저장하기
import numpy
# ict_수정.xls 파일을 읽어오기
df3 = pd.read_excel('./data/ict_수정.xlsx')
# 데이터프레임을 전치하기( 행열 위치 바꾸기)
df4 = df3.set_index('학과별(1)').T

answer = []
#df4[['전체계','전체ICT학과']].corr(numeric_only = True)

#corr_values = df4.corr().values
#corr_abs = numpy.abs(corr_values)
# print(corr_abs)

# for i in corr_abs:
#     for ii in i:
#         if(ii >= 0.5):
            



for i in df4.columns:
    for j in df4.columns:
        df5 = df4[[i,j]].corr(numeric_only = True)
        if( abs(df5.iloc[0,1]) > 0.5):
            answer.append(j)
answer_set = set(answer)
print(answer_set)
```
