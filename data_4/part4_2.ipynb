```python
#import
import matplotlib.pyplot as plt
import numpy as np # 수학적 함수 가지고 있는 라이브러리

plt.plot([1,2,3,4]) # y값 [1,2,3,4]
# x값을 생략하면 기본값을 [0,1,2,3,...] 으로 인식
plt.show()
```

```python
plt.plot([1,2,3,4],[1,9,4,16]) # plt.plot(x 값, y 값)
plt.show()
```

```python
plt.plot([1,2,3,4], [1,4,9,16], 'ro')
#           x축 값,   y축 값,   '포맷형식' red,마커형식 o
plt.axis([0,6,0,20])
# x축의 범위[xmin,xmax,ymin,ymax]
plt.show()
```

```python
t = np.arange(0., 5., 0.2) # 0. 부터 5.0 미만까지 0.2 간격으로 배열 만들어 냄
t
plt.plot(
    t, t**3, 'gs',
    t, t**2, 'b^',
    t, t, 'r-.',
    markersize = 4)
plt.show()
```

```python
# plot() 선이나 마커 그래프 그리는 함수
plt.plot([2,3,5,10])
plt.show()
```

```python
data_dict = {
    'data_x' : [1,2,3,4,5], 
    'data_y' : [2,3,5,10,8]
            }
plt.plot(data_dict['data_x'],data_dict['data_y']) # 원하는 딕셔너리를 유저가 직접 입력
plt.show()
```

```python
# 1. 기본 사용하기
plt.plot([1,4,9,16])
plt.xlabel('x-label')
plt.ylabel('y-label')
plt.show()
```

```python
# 2. 여백 지정하기
plt.plot([1,4,9,16])
plt.xlabel('x-label', labelpad = 15) # 15pt 라벨 여백 : x축 눈금과 x-label 글자 사이 간격
plt.ylabel('y-label', labelpad = 20) # 20pt 라벨 여백 : y축 눈금과 x-label 글자 사이 간격
plt.show()
```

```python
# 3. 폰트 설정하기
font1 = {
    'family' : 'serif',
    'color' : 'b',
    'weight' : 'bold',
    'size' : 14
}
plt.plot([1,4,9,16])
plt.xlabel('x-label', labelpad = 15, fontdict = font1)
plt.ylabel('y-label', labelpad = 20, fontdict = font1)
plt.show()
```

```python
# 4. 위치 지정하기
font1 = {
    'family' : 'serif',
    'color' : 'b',
    'weight' : 'bold',
    'size' : 14
}
plt.plot([1,4,9,16])
plt.xlabel('x-label', labelpad = 15, fontdict = font1, loc = 'right')
plt.ylabel('y-label', labelpad = 20, loc = 'top')
plt.show()
```

```python
# 범례 표시하기
# legend() 데이터 종류 표시하는 것
# keyword : legend(),label,loc,ncol,fontsize,frameon,shadow
# 기본 사용, 위치지정, 열 개수 지정, 폰트크기 지정, 범례 테두리 꾸미기
# 1. 기본 사용

data = {
    'x' : [1,2,3,4],
    'y' : [2,3,5,10]
}
plt.plot(
    data['x'], data['y'],
    linestyle = '-',
    label = 'line1'
)
plt.xlabel('x-Axis', labelpad = 15)
plt.ylabel('y-Axis', labelpad = 20)
plt.legend()
plt.show()
```

```python
# 2. 위치지정
data = {
    'x' : [1,2,3,4],
    'y' : [2,3,5,10]
}
plt.plot(
    data['x'], data['y'],
    linestyle = '-',
    label = 'line1'
)
plt.xlabel('x-Axis', labelpad = 15)
plt.ylabel('y-Axis', labelpad = 20)
# plt.legend(loc = 'lower left')
plt.legend(loc = (0.5,0.5))
plt.show()
```

```python
# 3. 열 개수 지정
data1 = {
    'x' : [1,2,3,4],
    'y' : [2,3,5,10]
}
data2 = {
    'x' : [1,2,3,4],
    'y' : [3,5,9,7]
}

plt.plot(
    data1['x'], data1['y'],
    'r-.',
    label = 'Price($)'
)
plt.plot(
    data2['x'], data2['y'],
    linestyle = '-',
    label = 'Demand(#)'
)

plt.xlabel('x-Axis', labelpad = 15)
plt.ylabel('y-Axis', labelpad = 20)
plt.legend(loc = (0.48,0.02), ncol = 2)
plt.show()
```

```python
# 4. 폰트크기 지정
data1 = {
    'x' : [1,2,3,4],
    'y' : [2,3,5,10]
}
data2 = {
    'x' : [1,2,3,4],
    'y' : [3,5,9,7]
}

plt.plot(
    data1['x'], data1['y'],
    'r-.',
    label = 'Price($)'
)
plt.plot(
    data2['x'], data2['y'],
    linestyle = '-',
    label = 'Demand(#)'
)

plt.xlabel('x-Axis', labelpad = 15)
plt.ylabel('y-Axis', labelpad = 20)
plt.legend(loc = (0.55,0.04), ncol = 2, fontsize = 8)
plt.show()
```

```python
# 5. 범례 테두리 꾸미기
data1 = {
    'x' : [1,2,3,4],
    'y' : [2,3,5,10]
}
data2 = {
    'x' : [1,2,3,4],
    'y' : [3,5,9,7]
}

plt.plot(
    data1['x'], data1['y'],
    'r-.',
    label = 'Price($)'
)
plt.plot(
    data2['x'], data2['y'],
    linestyle = '-',
    label = 'Demand(#)'
)

plt.xlabel('x-Axis', labelpad = 15)
plt.ylabel('y-Axis', labelpad = 20)
plt.legend(loc = (0.58,0.03), ncol = 2, fontsize = 8, frameon = False, shadow = True) # frameon은 default가 True, shadow는 Frameon이 True일 때만 작동
plt.show()
```

```python
# 2-1. 선 종류 변경
# linestyle, Solid '-', Dashed '--', Dotted ':', Dash-dot'-.'
plt.plot( [4,4,4],'r-.',label = 'Solid')
plt.plot( [3,3,3],'b--.',label = 'Dashed')
plt.plot( [2,2,2],'g:',label = 'Dotted')
plt.plot( [1,1,1], color = 'pink', marker = '*', linestyle = '-.', label = 'Dash-dot')

plt.axis([0.8,3.2,0.5,5.0])
plt.legend(loc = 'upper right', ncol = 4)

plt.show()
```

```python
# 2-2. 선 종류 변경
# linestyle, Solid '-', Dashed '--', Dotted ':', Dash-dot'-.'
plt.plot( [4,4,4], linestyle = 'solid', color = 'red', label = 'Solid')
plt.plot( [3,3,3], linestyle = 'dashed', color = 'blue',label = 'Dashed')
plt.plot( [2,2,2], linestyle = 'dotted', color = 'green',label = 'Dotted')
plt.plot( [1,1,1], linestyle = 'dashdot',color = 'hotpink', label = 'Dash-dot', marker = '*',)

plt.axis([0.8,3.2,0.5,5.0])
plt.legend(loc = 'upper right', ncol = 4)

plt.show()
```

```python
# 그래프 영역 채우기
# 그래프의 특정 영역을 채워서 강조할 때
# fill_between() - 두 수평 방향의 그래프 사이를 채운다
# fill_betweenx() - 두 수직 방향의 그래프 사이를 채운다
# fill() - 다각형 영역을 채운다
```

```python
# 수평 방향(x)기준의 그래프 영역 채우기
x = [1,2,3,4]
y = [2,3,5,10]

plt.plot(x,y, marker = 'p',color = 'navy')
plt.fill_between(x[2:4], y[2:4], alpha = 0.5)

plt.show()
```

```python
# 수직 방향(y)기준의 그래프 영역 채우기
x = [1,2,3,4]
y = [2,3,5,10]

plt.plot(x,y, marker = 'p',color = 'navy')
plt.fill_betweenx(y[2:4], x[2:4], alpha = 0.5)

plt.show()
```

```python
# 그래프 영역 채우기
x = [1,2,3,4]
y = [2,3,5,10]

plt.plot(x,y, marker = 'p',color = 'navy')
plt.fill_betweenx(y[2:4], x[2:4], alpha = 0.5)
plt.fill_between(x[2:4], y[2:4], alpha = 0.5)

plt.show()
```

```python
# 그래프 영역 채우기
x = [1,2,3,4]
y1 = [2,3,5,10]
y2 = [1,2,4,8]

plt.plot(x,y1)
plt.plot(x,y2)

plt.fill_between(x[1:3], y1[1:3], y2[1:3], color = 'lightgray', alpha = 0.5)

plt.show()
```
