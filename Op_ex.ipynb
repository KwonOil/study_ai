# 문제5) pandas의 연산 메소드(add(), sub(), mul(), div())를 활용하는 문제

상품 가격과 할인율이 아래처럼 Series로 주어졌습니다.



요구사항

연산 메소드 mul()과 sub()를 사용해서 아래 계산을 수행하세요.

최종가격 = 가격 - (가격 × 할인율/100)


조건:

mul()을 사용해 할인액을 계산.

sub()을 사용해 최종 가격을 계산.

최종 결과를 출력하세요.

```python
goods_name = ['키보드','마우스','모니터']
price = pd.Series([30000, 15000, 200000], index = goods_name)
discount = pd.Series([10, 5, 20], index = goods_name)

# print(goods_name)
print(price)
print(discount)

final_price = price.sub((price.mul(discount) / 100))
print(final_price)

df5 = pd.DataFrame([price, discount, final_price], index = ['가격', '할인율(%)', '최종가격']).T
df5
```

# 문제10) query() 메소드를 이용한 필터링 문제입니다

​

가격 범위로 상품 필터링하기

다음은 상품 가격과 관련된 데이터프레임입니다.



요구사항

가격이 30,000원 이상 100,000원 이하인 상품만 필터링하여 출력하세요.

```python
data10 = {
    '상품명' : ['키보드','마우스','모니터','헤드셋','웹캠'],
    '가격' : [30000, 15000, 200000, 80000, 120000]
}

df10 = pd.DataFrame(data10)
df10 = df10.set_index('상품명')
df10 = df10.query('30000 <= 가격 and 가격 <= 100000')

df10 = df10.rename(columns = {'가격':'가격(원)'})
df10
```

# 문제11) isin() 메소드를 이용한 문제입니다

isin()은 특정 값들이 존재하는지 확인하고 필터링할 때 정말 유용한 메소드입니다

특정 상품들만 필터링하기

다음은 다양한 상품들의 데이터입니다.



문제:

'키보드', '마우스', '웹캠' 상품만 필터링하여 출력하세요.

isin() 메소드를 활용해 필터링할 수 있습니다.

```python
data11= {
    '상품명' : ['키보드','마우스','모니터','헤드셋','웹캠','이어폰'],
    '가격' : [30000, 15000, 200000, 80000, 120000, 25000]
}
df11 = pd.DataFrame(data11)
print(df11)

isin11 = df11['상품명'].isin(['키보드','마우스','웹캠'])
df11 = df11[isin11]
df11
```
