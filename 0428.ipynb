```python
# 산점도 = 두 변수(열, 변량, 컬럼, 항목) 관계성
df = pd.read_csv('./data/auto-mpg.csv', header = None)
df.columns = ['mpg','cylinders','displacement','horsepower','weight','acceleration','model year','origin','name']

print(df.corr(numeric_only = True))
#print(df)
df.plot(x = 'weight', y = 'displacement', kind = 'scatter');
```

```python
# 박스플롯 - 데이터 분포, 분산정도
# 'o' - 이상치
# 사분위수에 있는 값의 분포
print(df[['mpg','cylinders']].mean(),'\n', df[['mpg','cylinders']].max(),'\n', df[['mpg','cylinders']].min())
print(df[['mpg','cylinders']].quantile(0.25)) # 데이터의 25% 지점의 값
print(df[['mpg','cylinders']].quantile(0.75)) # 데이터의 75% 지점의 값
df[['mpg','cylinders']].plot(kind = 'box');
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Jupyter Notebook이라면 그래프 바로 보이게:
%matplotlib inline
# 예시 데이터
np.random.seed(0)
df = pd.DataFrame({
    '월': ['1월', '2월', '3월', '4월'],
    '매출': [250, 270, 300, 280],
    '지출': [200, 230, 250, 240]
})

df_random = pd.DataFrame({
    '점수': np.random.normal(loc=70, scale=10, size=100),
    '키': np.random.normal(170, 10, 100),
    '몸무게': np.random.normal(65, 15, 100)
})
```

```python
df.plot(x='월', y=['매출', '지출'], kind='bar')
plt.title('월별 매출과 지출 비교')
plt.ylabel('금액')
plt.show()
```

```python
df.plot(x='월', y=['매출', '지출'], kind='barh')
plt.title('월별 매출과 지출 (수평)')
plt.xlabel('금액')
plt.show()
```

```python
df_random['점수'].plot(kind='hist', bins=15, alpha=0.7)
plt.title('점수 분포')
plt.xlabel('점수')
plt.show()
```

```python
df_random[['점수']].plot(kind='box')
plt.title('점수 박스플롯')
plt.show()
```

```python
df_random['점수'].plot(kind='kde')
plt.title('점수 커널 밀도 추정')
plt.xlabel('점수')
plt.show()
```

```python
df.set_index('월')[['매출', '지출']].plot(kind='area', alpha=1)
plt.title('면적 그래프 예시')
plt.ylabel('금액')
plt.show()
```

```python
df_sum = df[['매출', '지출']].sum()
df_sum.plot(kind='pie', autopct='%1.1f%%', startangle=90)
plt.title('전체 매출/지출 비율')
plt.ylabel('')
plt.show()
```

```python
df_random.plot(kind='scatter', x='키', y='몸무게', alpha=0.5)
plt.title('키 vs 몸무게 산점도')
plt.show()
```

```python
df_random.plot(kind='hexbin', x='키', y='몸무게', gridsize=25, cmap='viridis')
plt.title('키 vs 몸무게 고밀도 산점도')
plt.show()
```
