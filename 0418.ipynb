```python
# ====================== 데이터프레임 행/열 선택 =====================

data5 = {
         '주문번호' : [1001,1002,1003],
         '고객명' : ['김철수','이영희','박민수'],
         '상품명' : ['키보드','모니터','마우스'],
         '수량' : [2,1,3],
         '가격' : [30000,150000,200000]
        }

df5 = pd.DataFrame(data5)
df5 = df5.set_index(['주문번호'])

# 행 선택 하나만 할 때
# 변수 = 데이터프레임.loc[행 인덱스] 
# 행 선택 여러개 할 때
# 변수 = 데이터프레임.loc[[행 인덱스, 행 인덱스]] 
# 행 선택 연속되게 할 때
# 변수 = 데이터프레임.loc[시작 행 인덱스 : 마지막 행 인덱스]
label5 = df5.loc[1002]
print(label5)
# label5_2 = df5.loc[['수량','가격']]
# print(label5_2)

# 행 선택 하나만 할 때
# 변수 = 데이터프레임.iloc[정수형 인덱스]
# 행 선택 여러개 할 때
# 변수 = 데이터프레임.iloc[[정수형 인덱스, 정수형 인덱스]] 
# 행 선택 연속되게 할 때
# 변수 = 데이터프레임.iloc[시작 정수형 인덱스 : 마지막 정수형 인덱스]
position5 = df5.iloc[0:2]

# 열 선택 하나만 할 때
# 변수 = 데이터프레임[열 인덱스] or 데이터프레임.열 인덱스 
# 열 선택 여러개 할 때
# 변수 = 데이터프레임[[열 인덱스, 열 인덱스]]
# 열 선택 연속되게 할 때
# 변수 = 데이터프레임[시작 열 인덱스 : 마지막 열 인덱스]
answer5 = position5[['고객명', '상품명']]
answer5
```

# 문제6)  원소값 수정하기 

학생 성적 데이터 수정하기

다음 데이터를 DataFrame으로 만들고,

특정 원소 값을 수정하세요!

요구사항

- 영희의 수학 점수를 100 → 98로 변경하세요.

- 민수의 영어 점수를 80 → 82로 수정하세요.

```python
# ====================== 데이터프레임 원소 선택/변경 =====================
data6 = {
         '이름' : ['철수','영희','민수'],
         '국어' : [90,70,85],
         '영어' : [85,95,80],
         '수학' : [80,100,90]
        }

df6 = pd.DataFrame(data6)

df6 = df6.set_index('이름')
print(df6)
print()

# 같은 행에서 하나의 원소 찾기
# 데이터프레임.loc[행 인덱스, 열 인덱스] <-- loc로 찾기
# 데이터프레임.loc[정수형 행 인덱스, 정수형 열 인덱스] <-- iloc로 찾기
# 위 선택된 원소에 새로운 값을 대입하면 변경
df6.loc['영희','수학'] = 98
df6.loc['민수','영어'] = 82
print(df6)
print()

# 같은 행에서 여러 개의 원소 찾기
# 데이터프레임.loc[행 인덱스, [열 인덱스, 열 인덱스, ...]] <-- loc로 찾기
# 데이터프레임.loc[정수형 행 인덱스, [정수형 열 인덱스, 정수형 열 인덱스, ...]] <-- iloc로 찾기
# 위 선택된 원소에 새로운 값을 대입하면 변경
df6.loc['영희',['영어','수학']] = 100, 100
df6.iloc[2,[0,1]] = 100, 100
print(df6)
print()
```

# 문제8) reindex() 행 인덱스 재배열

​

*** reindex() 주로 사용하는 분야

데이터 정렬 변경

원래 순서와 관계없이, 원하는 순서로 인덱스를 다시 배열할 때.

누락된 인덱스 추가 (NaN)

데이터에 없는 값도 인덱스로 추가해서 결측치를 생성, 이후 fillna로 처리 가능.

다른 DataFrame과 인덱스를 맞출 때

다른 데이터셋과 병합하거나 비교할 때, 인덱스를 동일하게 맞출 수 있음.


---
문제 

아래 데이터를 DataFrame으로 만든 뒤,

reindex()를 사용해 주문번호 순서를 [1003, 1001, 1004, 1002]로 변경하세요.



참고 : 1004 주문번호는 원래 데이터에 없으므로 NaN 값이 나타나야 합니다.

​

결과화면



```python
# ====================== 데이터프레임 행 재배열 =====================
data8 = {
         '주문번호' : [1001,1002,1003],
         '고객명' : ['김철수','이영희','박민수'],
         '상품명' : ['키보드','모니터','마우스']
        }

df8 = pd.DataFrame(data8)
df8 = df8.set_index(['주문번호'])

# 새로운 인덱스 리스트
new_index = [1003,1001,1004,1002]
# df8의 인덱스는 new_index로 새로 설정하고 df8에 대입 
df8 = df8.reindex(new_index)
# new_index의 개수가 원래 개수보다 하나 더 많아 행이 추가됨
# 새로운 행에는 NaN값이 기본값으로 들어간다 

df8
```

중복된 인덱스가 있을 때 재정비

```python
student1 = pd.Series({'국어': 100,
                       '영어': 80,
                       '수학': 90,
                     })
student2 = pd.Series({'국어': 80,
                       '영어': 90,
                       '수학': 80,
                     })

student_p = student1 + student2
student_s = student1 - student2
student_m = student1 * student2
student_d = student1 / student2

result = pd.DataFrame([student_p,student_s,student_m,student_d],index = ['덧셈','뺄셈','곱셈','나눗셈']).T
result
```

```python
# p62 예제 1-24
student1 = pd.Series({'국어': np.nan,
                       '영어': 80,
                       '수학': 90,
                     })
student2 = pd.Series({'수학': 80,
                       '국어': 90
                     })

sr_add = student1.add(student2, fill_value = 0)
sr_sub = student1.sub(student2, fill_value = 0)
sr_mul = student1.mul(student2, fill_value = 0)
sr_div = student1.div(student2, fill_value = 0)

result = pd.DataFrame([sr_add, sr_sub, sr_mul, sr_div], index = ['덧셈','뺄셈','곱셈','나눗셈'])
result
```
